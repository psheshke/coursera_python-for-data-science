```python
import datetime
import time
```

```python
df['end_hour'] = df['stoptime'].apply(lambda x: datetime.datetime.fromtimestamp(
            time.mktime(datetime.datetime.strptime(x.strip(), "%Y-%m-%d %H:%M:%S.%f").timetuple())).hour)
```

```python
df['start_hour'] = df['starttime'].apply(lambda x: datetime.datetime.fromtimestamp(
                                        time.mktime(datetime.datetime.strptime(x.strip(), "%Y-%m-%d %H:%M:%S.%f").timetuple())).hour)
```

```python
df[(df.start_hour.isin([18, 19, 20]))]['start station id'].value_counts().nlargest(1)
```

```python
df[(df.end_hour.isin([6, 7, 8, 9, 10])) & (df['end station id'].isin([3140, 3106, 3116, 369]))]['end station id'].value_counts()
```
